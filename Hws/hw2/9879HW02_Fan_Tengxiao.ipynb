## The Parlour (1998) model

### 1. (10 points) 
Consider the first step $t = 1$ of the 3-step order book case presented in the lecture slides. Suppose that the order book at (or just before) $t = 1$ has one share offered and one share bid ($n_B = n_A = 1$).

Determine the probabilities of a market buy, market sell, limit buy and limit sell at $t = 1$.

In the lecture slides, the parameters of the 3-step order book case is:
$$
V=5.5;\,B=5;\,A=6; \,T=3 \text { and } \beta_t \sim U(0,2).
$$ 
Where $V$ is the future price, $B$ is the bid price, $A$ is the ask price and $\beta_t$ is the random discount parameter. Thus, we have:
- Market sell (MS) payoff: $B - \beta_t\,V$
- Limit sell (LS) payoff: $(A - \beta_t\,V)\,{\mathbb{P} }^B_{t+i}$
- Limit buy (LB) payoff: $(\beta_t\,V-B)\,{\mathbb{P} }^S_{t+i}$
- Market buy (MB) payoff: $\beta_t\,V-A$  

Due to symmetric, we only consider market sell and limit sell. First, we calculate the probability that a limited sell order will be executed in the future if it is placed. When a limited sell order is placed, there will be two share offered in the order book. Thus, there will be no limited sell order in period 2 and 3, and the order we placed will only be executed if there are market buy orders in both period 2 and 3. The probability is:
$${\mathbb{P} }^B_{1+i}={\mathbb{P} }^B_{2}\times {\mathbb{P} }^B_{3}=\left(\frac{5}{22}\right)^2=\frac{25}{484}$$
Thus, submit a MS only if
$B - \beta_1\,V \geq (A - \beta_1\,V) \,{\mathbb{P} }^B_{1+i}$ or
equivalently
$$\beta_1 \leq \frac{B-A\,{\mathbb{P} }^B_{1+i}}{V\,(1-{\mathbb{P} }^B_{1+i})} = \frac{B-A\,{\mathbb{P} }^B_{1+i}}{V\,(1-{\mathbb{P} }^B_{1+i})} =\frac{4540}{5049}.$$
As a result, a seller will choose as follows:

- Market sell (MS) if $B - \beta_1 V \geq (A-\beta_t V)^+ {\mathbb{P} }^B_{1+i}$, i.e., $0 \leq \beta_1 < \frac{4540}{5049}$
- Limit sell(LS) if $(A-\beta_t\,V)\,{\mathbb{P} }^B_{1+i}>(B-\beta_t\,V)^+$, i.e., $\frac{4540}{5049} \leq \beta_1<\frac{12}{11}$
- Do nothing if $\frac{12}{11}\leq \beta_1 \leq 2$<br>
Since the probability of a buyer and a seller come to the market is both $\frac{1}{2}$, we have the results:
\begin{eqnarray*}
{\mathbb{P} }_{MB}&=&{\mathbb{P} }_{MS}&=&\frac{1}{2}{\mathbb{P} }\left(\beta_1<\frac{4540}{5049}\right)=\frac{1135}{5049}\\
{\mathbb{P} }_{LB}&=&{\mathbb{P} }_{LS}&=&\frac{1}{2}{\mathbb{P} }\left(\frac{4540}{5049}\leq\beta_1<\frac{12}{11}\right)=\frac{22}{459}\\
{\mathbb{P} }_{0}&=&\frac{5}{11}
\end{eqnarray*}

In which, ${\mathbb{P} }_{0}$ includes the case of doing nothing (either seller or buyer). The probability of a seller or a buyer doing nothing is $\frac{5}{22}$.

### 2. (4 points) 
Explain why the optimal market/limit order decision depends in general on quantity available on both sides of the order book in the Parlour model.

For simplicity, let us consider buy orders.<br>
If the quantity on bid side increases, limit buy orders will decrease because the order will be filled after all the other bid side orders are filled so that the probability of the order being filled decreases. And the market buy orders will increase in contrast.<br>
If the quantity on ask side decreases, the probability of a market sell order in the subsequent periods will decrease (because the limit sell orders will increase since it's more likely to be filled). Thus, the probability of a limit buy order will decrease because the probability of it being filled by future market sell order decreases. And the market buy orders will increase in contrast.<br>
It is the same case for sell orders.

## The Bouchaud-M&eacute;zard-Potters order book approximation

### 4. (15 points)

When $\lambda(u)=\lambda$,  a constant, the BMP order book density becomes

\begin{eqnarray*}
\rho_{BMP}(\Delta) &=& \frac{\lambda}{\nu}\,\left\{e^{-\Delta/{\tilde \Delta}}\,\int_0^{\Delta/{\tilde\Delta}}\,\sinh(u)\,du + \sinh(\Delta/{\tilde \Delta})\,\int_{\Delta/{\tilde\Delta}}^\infty\,e^{-u}\,du\right\} \\
&=&\frac{\lambda}{\nu}\,\left\{1 - e^{-\Delta/{\tilde \Delta}}  \right\},
\end{eqnarray*}

where the characteristic distance $\tilde\Delta$ is given by

$$
\tilde \Delta=\sqrt{\frac{\sigma^2}{2\,\nu}}.
$$

(b)<br>
Parameters are $\lambda=1,\mu=10,\nu=\frac{1}{5}$
$$\sigma_{DA}^2=\frac{\mu^2\nu}{\lambda^2}=\frac{10^2*1/5}{1^2}=20$$
$$\tilde{\Delta}_{DA}=\sqrt{\frac{\sigma_{DA}^2}{2*\nu}}=\sqrt{20/(2*1/5)}=\sqrt{50}$$

(c)<br>
The density function is:
$$\rho_{DA}(\Delta)=\frac{\lambda}{\nu}\,\left\{1 - e^{-\Delta/{\tilde \Delta}}  \right\}=5(1-\exp(-\Delta/\sqrt{50}))$$

```python
def rhoDA (Delta, lmda, mu, nu):
    sigDA=mu*mu*nu/lmda/lmda
    char=np.sqrt(sigDA/2/nu)
    return lmda/nu*(1-np.exp(-np.abs(Delta)/char))
```

(e) Write a function with prefactor, which not only gives the scale.

```python
def rhoDA_prefactor (pre, Delta, lmda, mu, nu):
    sigDA=mu*mu*nu/lmda/lmda
    char=np.sqrt(sigDA/2/nu)
    return pre*lmda/nu*(1-np.exp(-np.abs(Delta)/char))
```

(g)<br>
Take the minimum square error as the criteria of the best prefactor. From the graph we can see that the prefactor should be larger than 1. We try to find 

```python
# Function to calculate square error
def sqr_error(avg_book_shape, rho):
    err=0
    for i in range(len(rho)):
        err += (avg_book_shape[i]-rho[i])**2
    return err

# Find the smallest error
prefactor = 1
rho = [rhoDA_prefactor(prefactor, i, lmda, mu, nu) for i in x]
former_err = sqr_error(avg_book_shape , rho)
while True:
    prefactor += 0.01
    rho = [rhoDA_prefactor(prefactor, i, lmda, mu, nu) for i in x]
    err = sqr_error(avg_book_shape , rho)
    if err > former_err:
        print ("Prefactor:", prefactor-0.01, "Error: ", former_err)
        break
    former_err = err
```
