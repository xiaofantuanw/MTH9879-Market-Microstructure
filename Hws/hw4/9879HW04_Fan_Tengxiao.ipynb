## Glosten and Milgrom

### 1. (8 points) 

(Adapted from Hasbrouck 5.2 but please note that the setup in (3) in the slides is different from Hasbrouck's in Figure 5.1 of his book.) Consider a variant of the Glosten and Milgrom model where there is informed trading only in the low state $V=\underline V$.

(a) Show that the learned probability after one step that the true price is $\bar V$ is given by

$$
\delta_1(Buy)=\Pr(\bar V|Buy)=\frac{\delta}{1-\mu\,(1-\delta)};\quad \delta_1(Sell)=\Pr(\bar V|Sell)=\frac{\delta}{1+\mu\,(1-\delta)}
$$

and that

$$
A=\frac{\underline V\,{(1-\delta)\,(1-\mu)}+\bar V\,{\delta}}{{1-\mu\,(1-\delta)}};\quad B=\frac{\underline V\,{(1-\delta)\,(1+\mu)}+\bar V\,{\delta}}{1+\mu\,(1-\delta)}
$$

(b) Deduce an expression for the spread $s=A-B$.

(c) Explain the sensitivities of the spread to the parameters of the model to a non-expert.


#### <font color=blue> Solution 1(a). </font>

The unconditioned probability of Buy and Sell is:
$$\Pr(Buy)=\frac{1}{2}\delta+\frac{1}{2}(1-\delta)(1-\mu)=\frac{1}{2}(1-\mu(1-\delta))$$
$$\Pr(Sell)=\frac{1}{2}\delta+\frac{1}{2}(1-\delta)(1-\mu)+(1-\delta)\mu=\frac{1}{2}(1+\mu(1-\delta))$$
The probability of price going up with a buy/sell order is
$$\Pr(\bar{V},Buy)=\Pr(\bar{V},Sell)=\frac{1}{2}\delta$$
Thus the conditional probability (i.e. The learned probability after one step) can be expressed as 
$$\delta_1(Buy)=\Pr(\bar{V}|Buy)=\frac{\Pr(\bar{V},Buy)}{\Pr(Buy)}=\frac{\delta}{1-\mu(1-\delta)}$$
$$\delta_1(Sell)=\Pr(\bar{V}|Sell)=\frac{\Pr(\bar{V},Sell)}{\Pr(Sell)}=\frac{\delta}{1+\mu(1-\delta)}$$
Similarly, probability of price going down with a buy/sell order is
$$\Pr(\underline{V},Buy)=\frac{1}{2}(1-\delta)(1-\mu)$$
$$\Pr(\underline{V},Sell)=\frac{1}{2}(1-\delta)(1-\mu)+(1-\delta)\mu$$
Thus the conditional probability is
$$\Pr(\underline{V}|Buy)=\frac{\Pr(\underline{V},Buy)}{\Pr(Buy)}=\frac{(1-\delta)(1-\mu)}{1-\mu(1-\delta)}$$
$$\Pr(\underline{V}|Sell)=\frac{\Pr(\underline{V},Sell)}{\Pr(Sell)}=\frac{(1-\delta)(1+\mu)}{1-\mu(1-\delta)}$$
Finally we get the bid and ask prices
$$A=\E[V|Buy]=\bar{V}\Pr(\bar{V}|Buy)+\underline{V}\Pr(\underline{V}|Buy)=\frac{\underline V\,{(1-\delta)\,(1-\mu)}+\bar V\,{\delta}}{{1-\mu\,(1-\delta)}}$$
$$B=\E[V|Sell]=\bar{V}\Pr(\bar{V}|Sell)+\underline{V}\Pr(\underline{V}|Sell)=\frac{\underline V\,{(1-\delta)\,(1+\mu)}+\bar V\,{\delta}}{1+\mu\,(1-\delta)}$$

#### <font color=blue> Solution 1(b). </font>

The spread is the difference between ask and bid prices:
$$s=A-B=\frac{\underline V\,{(1-\delta)\,(1-\mu)}+\bar V\,{\delta}}{{1-\mu\,(1-\delta)}}-\frac{\underline V\,{(1-\delta)\,(1+\mu)}+\bar V\,{\delta}}{1+\mu\,(1-\delta)}=\frac{2\mu\delta(1-\delta)(\bar{V}-\underline{V})}{1-\mu^2(1-\delta)^2}$$

#### <font color=blue> Solution 1(c). </font>

When $\mu$ increases, the spread will increase. This is because when there are more informed traders, it is harder for the market maker to make money, thus market makers will increase the spread to buy at a lower price and sell at a higher price. In extreme cases, when $\mu=1$ which means all the traders are informed, the spread will be $2\delta(1-\delta)(\bar{V}-\underline{V})$, and the result is $\bar{V}-\underline{V}$ when $\delta=\frac{1}{2}$, meaning that there will be no orders quoted.<br>
When $\bar{V}-\underline{V}$, meaning that the market is more volatile (changes faster), the market maker will increase the spread to compensate its rising risk.<br>
When $\delta$ increases, which means the probability of price rising to $\bar{V}$ in the next period increases, the spread will decrease, because the probability of a trade made by informed traders will decrease due to less probability of price going down.

## Predicting trade signs

### 4. (6 points) 

The following require the `trade_signs` series from Problem 2.

(a) Denote the probability of a sign reversal by $\pi$.  Derive an expression in terms of $\pi$ for the conditional expectation $\E[\epsilon_t|\epsilon_{t-1}]$.

(b) Assume zero expectation of the trading signs, fit an AR(1) model to the series `trade_signs`.  What is the coefficient returned by this model?  Explain how  this coefficient relates to $\pi$?


#### <font color=blue> Solution 4(a). </font>

The probability of a sign reversal is
$$\mathbb{P}(\epsilon_t=-\epsilon_{t-1})=\pi$$
The probability of having a same sign is
$$\mathbb{P}(\epsilon_t=\epsilon_{t-1})=1-\pi$$
Thus the conditional expectation is
$$\E[\epsilon_t|\epsilon_{t-1}]=(1-\pi)\epsilon_{t-1}-\pi\epsilon_{t-1}=(1-2\pi)\epsilon_{t-1}$$

#### <font color=blue> Solution 4(b). </font>

Use the following Python code:

```python
from statsmodels.tsa.ar_model import AutoReg, ar_select_order
```

Since the trade signs are zero mean, the auto regression model has no intercept. $$\epsilon_t=k\epsilon_{t-1}$$
Thus $k=1-2\pi$, we can get
$$\pi=\frac{1}{2}(1-k)$$

```python
# Fit an AR(1) model to the trade signs series without an intercept
trade_signs = tqdata['SIGN'].values
model = AutoReg(trade_signs, lags=1, trend='n')  # 'n' indicates no trend/intercept
model_fitted = model.fit()

# Print the summary of the model
print(model_fitted.summary())
```

```python
k = model_fitted.params[0]
pi = (1-k)/2
print("The coefficient of AR(1) model is",k)
print("The conversion rate is",pi)
```

Using the estimate of $\rho$ (the coefficient obtained from fitting to an AR(1) model) from Problem 4 and equation (11) from the lecture slides:

$$
\Delta p_t=(\phi+\lambda)\,\epsilon_t-(\phi+\rho\,\lambda)\,\epsilon_{t-1}+e_t+\Delta \xi_t

estimate by linear regression the MRR parameters $\phi$ (transaction cost) and $\lambda$ (adverse selection or "Kyle lambda") from  `tqdata`.  This amounts to ignoring the autocorrelation of the $\Delta \xi_t$.

Hint: Use the `ols` method in the `Python` module `statsmodels.formula.api` to perform the linear regressions.

```python
# Import statsmodels module for linear regression
import statsmodels.formula.api as sm
```

#### <font color=blue> Solution 5. </font>

From the equation (11), we can get the relationship between parameters of the regression and the parameters to solve.
$$\beta_1=\phi+\lambda$$$$\beta_2=-\phi-\rho\lambda$$
Then we solve the equations and get
$$\lambda=\frac{\beta_1+\beta_2}{1-\rho}$$
$$\phi=\frac{\rho\beta_1+\beta_2}{\rho-1}$$

```python
# Get the difference of prices and signs
prices = tqdata["PRICE"]
n=len(prices)
Delta_p = np.diff(prices)
epsilon_t_minus = trade_signs[0:n-1]
epsilon_t = trade_signs[1:n]
# Create a DataFrame from the arrays
df = pd.DataFrame({
    'Delta_p': Delta_p,
    'epsilon_t_minus': epsilon_t_minus,
    'epsilon_t': epsilon_t,
})

# Construct the regression formula
formula = 'Delta_p ~ epsilon_t + epsilon_t_minus'

# Perform the linear regression using OLS
results = sm.ols(formula, data=df).fit()

print(results.summary())
```
