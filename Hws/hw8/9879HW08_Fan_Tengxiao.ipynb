## Predictability of order flow

### 1. (6 points)

Suppose that the distribution of meta order sizes $Q$ is exponential so that

$$
\Pr(Q>V)= \frac{1}{\lambda}\,e^{-\lambda\,V}
$$

Assuming slices of  constant size $\bar v$, compute the conditional probability that a hidden order has more slices given that $n$ slices have already been observed.  Compare your result with the case where distribution of metaorder size is power-law.

#### <font color=blue> Solution 1. </font>

The probability that a hidden order has more slices given $n$ slices have been observed is
$$\begin{align}\Pr(Q>(n+1)\bar{v}|Q>n\bar{v})&=\frac{\Pr(Q>(n+1)\bar{v})}{\Pr(Q>n\bar{v})}\\&=\frac{e^{-\lambda(n+1)\bar{v}}}{e^{-\lambda n\bar{v}}}\\&=e^{-\lambda\bar{v}}\end{align}$$
In the case where metaorder size is power-law, the conditional probability is
$$\Pr(Q \geq n+1 \mid Q \geq n)=\frac{\operatorname{Pr}(Q \geq n+1)}{\operatorname{Pr}(Q \geq n)} \sim\left(\frac{n}{n+1}\right)^\beta$$
We can see that as the meta order size increases, the probability that the order continues to come remains unchanged in the exponential distribution case, but in the power-law case, the probability increases.

## Price manipulation

### 2. (10 points)

Consider the price process

$$
S_t=S_0+\int_0^t\,f(\dot{x}_s)\,G(t-s)\,ds+\int_0^t\,\sigma\,dZ_s
$$

where ${\dot x}_s$ is the rate of trading in dollars at time $s < t$, $f(\dot{x}_s)$ represents the impact of trading at time $s$ and $G(t-s)$ is a decay factor.

- (a) With the choices
$$
f(v)=\sqrt{v},\quad G(\tau)=\frac{1}{1+\tau},
$$
repeat the computations in Lecture 8  to derive an expression for the expected cost of accumulating shares at the rate $v_1$ and liquidating them at the rate $v_2$.

- (b) Verify that with $T=1$, $v_1=1/4$, $v_2=1$, the expected cost of this round-trip strategy is negative, thus demonstrating that price manipulation is possible. (Recall from the lecture slides that price manipulation is always possible if $f(\cdot)$ is nonlinear and the decay kernel $G(\cdot)$  is finite at the origin.)

Hint: You might consider using Mathematica or Wolfram Alpha to do the integrations.

#### <font color=blue> Solution 2(a). </font>

The cost of this trading strategy is 
$$C(\Pi)=C_{11}+C_{22}-C_{12}$$where
$$\begin{aligned} & C_{11}=v_1 f\left(v_1\right) \int_0^{\theta T} d t \int_0^t G(t-s) d s \\ & C_{22}=v_2 f\left(v_2\right) \int_{\theta T}^T d t \int_{\theta T}^t G(t-s) d s \\ & C_{12}=v_2 f\left(v_1\right) \int_{\theta T}^T d t \int_0^{\theta T} G(t-s) d s\end{aligned}$$
and $$\theta=\frac{v_2}{v_1+v_2}$$
Plug in $$
f(v)=\sqrt{v},\quad G(\tau)=\frac{1}{1+\tau},
$$
Thus
$$\begin{align}C(\Pi)&=C_{11}+C_{22}-C_{12}\\&=v_1\sqrt{v_1}\int_0^{\theta T}\log(1+t)dt+v_2\sqrt{v_2}\int_{\theta T}^{T}\log(1+t-\theta T)dt-v_2\sqrt{v_1}\int_{\theta T}^T(\log(1+t)-\log(1+t-\theta T))dt\\&=v_1 \sqrt{v_1}((1+\theta T)(\log (1+\theta T)-1)+1)+v_2 \sqrt{v_2}((1+T-\theta T)(\log (1+T-\theta T)-1)+1)\\
&-v_2 \sqrt{v_1}((1+T)(\log (1+T)-1)-(1+\theta T)(\log (1+\theta T)-1)-(1+T-\theta T)(\log (1+T-\theta T)-1)-1)
\end{align}$$

#### <font color=blue> Solution 2(b). </font>

Take $T=1,v_1=1/4,v_2=1$, we have $\theta=4/5$. Use Python to do the calculation and get:


```python
import sympy as sp

T = 1
v1 = 1/4
v2 = 1
theta = 4/5

result = v1 * sp.sqrt(v1) * ((1 + theta * T) * (sp.log(1 + theta * T) - 1) + 1) \
        + v2 * sp.sqrt(v2) * ((1 + T - theta * T) * (sp.log(1 + T - theta * T) - 1) + 1) \
        - v2 * sp.sqrt(v1) * ((1 + T) * (sp.log(1 + T) - 1) - (1 + theta * T) * (sp.log(1 + theta * T) - 1) \
        - (1 + T - theta * T) * (sp.log(1 + T - theta * T) - 1) - 1)

print("Expected cost of the round trip:" ,result.simplify())
```

Since the cost is negative, manipulation is possible by accumulating and liquidating the asset.

## The Obizhaeva-Wang model

### 3. (8 points)

In the Obizhaeva Wang (OW) model, the price follows

$$
S_t=S_0+\eta\,\int_0^t\,{\dot x}_s\,e^{-\rho\,(t-s)}\,ds+\int_0^t\,\sigma\,dZ_s.
$$

- (a) Derive an expression for the expected price path corresponding to a VWAP execution.
- (b) With the duration $T$ of the meta order set to 1, plot the expected price path from time $t=0$ to time $t=2$ for $\rho=0.5,1,$ and $2$.
- (c) What is the expected average cost per share $\bar S_T$ of a VWAP order in the OW model?

#### <font color=blue> Solution 3(a). </font>

For a VWAP execution, we do the order execution with a certain velocity $v$ until we finish the order. The velocity is
$$\dot{x}_t=\begin{cases}v,t\leq T\\0,t>T\end{cases}$$
Since the expectation of volatility term is 0, the expected price path is
$$E[S_t]=S_0+\eta\int_0^t {\dot x}_s\,e^{-\rho\,(t-s)}\,ds$$
When $t\leq T$, $$E[S_t]=S_0+\frac{\eta v}{\rho}\left(1-e^{-\rho t}\right)$$
When $t>T$, $$E[S_t]=S_0+\eta\int_0^T {\dot x}_s\,e^{-\rho\,(t-s)}\,ds=S_0+\frac{\eta v}{\rho}\left(e^{-\rho(t-T)}-e^{-\rho t}\right)$$
As a result,
$$E\left[S_t\right]=\left\{\begin{array}{l}S_0+\frac{\eta v}{\rho}\left(1-e^{-\rho t}\right), t \leq T \\ S_0+\frac{\eta v}{\rho}\left(e^{-\rho(t-T)}-e^{-\rho t}\right), t>T\end{array}\right.$$

#### <font color=blue> Solution 3(b). </font>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
S0 = 0  
eta = 0.1
v = 1 #We have no velocity requirement
T = 1
rho_values = [0.5, 1, 2]
t = np.linspace(0, 2, 1000)


def expected_st(t, rho):
    before_T = S0 + (eta * v / rho) * (1 - np.exp(-rho * t))
    after_T = S0 + (eta * v / rho) * (np.exp(-rho * (t - T)) - np.exp(-rho * t))
    return np.where(t <= T, before_T, after_T)


plt.figure(figsize=(10, 6))
for rho in rho_values:
    plt.plot(t, expected_st(t, rho), label=f'rho={rho}')

plt.title('Expected Price Path Over Time')
plt.xlabel('Time t')
plt.ylabel('Expected $S_t$ for $S_0=0$')
plt.legend()
plt.show()
```

#### <font color=blue> Solution 3(c). </font>

For VWAP execution, the velocity is meta order size divided by time
$v=\frac{N}{T}$.
The cost per share is total cost divided by shares $\bar{S}_T=\frac{C[\Pi]}{N}$<br>
As the total cost is the integral of the price path before time $T$, we have 
$$\begin{align} C[\Pi]&=\int_0^T \frac{N}{T} d t \int_0^t \eta \frac{N}{T} e^{-\rho(t-s)} d s\\&=\int_0^T\frac{\eta N^2}{\rho T^2}\left(1-e^{-\rho t}\right)dt\\&=\frac{\eta N^2}{\rho T}+\frac{\eta N^2}{\rho^2 T^2}(e^{-\rho T}-1)\end{align}$$
Thus the average cost per share is 
$$\bar{S}_T=\frac{C[\Pi]}{N}=\frac{\eta N}{\rho T}+\frac{\eta N}{\rho^2 T^2}(e^{-\rho T}-1)$$
